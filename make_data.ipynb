```python
import numpy as np
import pandas as pd
```

```python
def make_data_demands(t,m):#m是客户数，t是天数
    demand=np.random.randint(1000,50000,(t,m))
    data_demand=pd.DataFrame(demand)
    data_demand.columns=np.arange(1,m+1)
    data_demand.insert(0,'mtime',np.arange(1,t+1))
    data_demand.to_csv('demands.csv',index=False)
    return data_demand
```

```python
def make_site_bandwidths(n):
    bandwidth=np.random.randint(100000,1000000,n)
    data_bandwidths=pd.DataFrame()
    data_bandwidths['bandwidth']=bandwidth
    data_bandwidths.insert(0,'site_name',np.arange(1,n+1))
    data_bandwidths.to_csv('site_bandwidth.csv',index=False)
    return data_bandwidths
```

```python
def make_qos(n,m):
    qos=np.random.randint(100,700,(n,m))
    data_qos=pd.DataFrame(qos)
    data_qos.columns=np.arange(1,m+1)
    data_qos.insert(0,'site_name',np.arange(1,n+1))
    data_qos.to_csv('qos.csv',index=False)
    return data_qos
```

```python
# data_qos=make_qos(10,3)
# data_qos
```

```python
def make_data(n,m,t):
    make_data_demands(t,m)
    make_site_bandwidths(n)
    make_qos(n,m)
```
